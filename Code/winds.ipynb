```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from matplotlib import rcParams, rc
```

```python


# set rcParams ~~~~~~~~~~~~~~~~~~~~~~~~
rcParams['axes.linewidth'] =    2

rcParams['xtick.major.size'] =  10
rcParams['xtick.major.width'] = 3
rcParams['xtick.minor.size'] =  5
rcParams['xtick.minor.width'] = 3

rcParams['ytick.major.size'] =  10
rcParams['ytick.major.width'] = 3
rcParams['ytick.minor.size'] =  5
rcParams['ytick.minor.width'] = 2
font = {'size' : 16, 'weight' : 'bold'}
rc('font', **font)


#davos, devon, lapaz, oslo
cm_name = 'devon'
cm_file = np.loadtxt(f'/home/imalsky/Desktop/New_Jups/Code/ScientificColourMaps5/{cm_name}/{cm_name}.txt')
cm_file = np.flip(cm_file, axis=0)
my_colors = mcolors.LinearSegmentedColormap.from_list(cm_name, cm_file)
```

```python
df1 = pd.read_csv('/home/imalsky/Desktop/UPS-PLANETS/OG-GCM-OUTPUT/UPS-BIG-G-CLEAR.txt',
                 delim_whitespace=True, skiprows=5,
                 names=('lat', 'lon', 'level',
                        'alt', 'pres', 'temp', 
                        'u', 'v', 'w',
                        'aero_sw_tau_1', 'sw_asym_1', 'sw_pi0_1'))

df1 = df1[["lat", "level", "pres", "u"]]
df1 = df1.groupby(['lat', 'level']).mean().reset_index()
df1 = df1[["lat", "pres","u"]]

u_data_1 = df1.u.values.reshape(48, 50)







df2 = pd.read_csv('/home/imalsky/Desktop/UPS-PLANETS/OG-GCM-OUTPUT/UPS-LOW-G-CLEAR.txt',
                 delim_whitespace=True, skiprows=5,
                 names=('lat', 'lon', 'level',
                        'alt', 'pres', 'temp',
                        'u', 'v', 'w',
                        'aero_sw_tau_1', 'sw_asym_1', 'sw_pi0_1'))

df2 = df2[["lat", "level", "pres", "u"]]
df2 = df2.groupby(['lat', 'level']).mean().reset_index()
df2 = df2[["lat", "pres","u"]]

u_data_2 = df2.u.values.reshape(48, 50)









df3 = pd.read_csv('/home/imalsky/Desktop/UPS-PLANETS/OG-GCM-OUTPUT/UPS-LOW-G-EXT-CLOUDY.txt',
                 delim_whitespace=True, skiprows=5,
                 names=('lat', 'lon', 'level',
                        'alt', 'pres', 'temp', 'temp_std',
                        'u', 'v', 'w',
                        'aero_sw_tau_1', 'sw_asym_1', 'sw_pi0_1',
                        'aero_sw_tau_2', 'sw_asym_2', 'sw_pi0_2',
                        'aero_sw_tau_3', 'sw_asym_3', 'sw_pi0_3',
                        'aero_sw_tau_4', 'sw_asym_4', 'sw_pi0_4'))

df3 = df3[["lat", "level", "pres", "u"]]
df3 = df3.groupby(['lat', 'level']).mean().reset_index()
df3 = df3[["lat", "pres","u"]]

u_data_3 = df3.u.values.reshape(48, 50)












df4 = pd.read_csv('/home/imalsky/Desktop/UPS-PLANETS/OG-GCM-OUTPUT/UPS-LOW-G-COM-CLOUDY.txt',
                 delim_whitespace=True, skiprows=5,
                 names=('lat', 'lon', 'level',
                        'alt', 'pres', 'temp', 'temp_std',
                        'u', 'v', 'w',
                        'aero_sw_tau_1', 'sw_asym_1', 'sw_pi0_1',
                        'aero_sw_tau_2', 'sw_asym_2', 'sw_pi0_2',
                        'aero_sw_tau_3', 'sw_asym_3', 'sw_pi0_3',
                        'aero_sw_tau_4', 'sw_asym_4', 'sw_pi0_4'))

df4 = df4[["lat", "level", "pres", "u"]]
df4 = df4.groupby(['lat', 'level']).mean().reset_index()
df4 = df4[["lat", "pres","u"]]

u_data_4 = df4.u.values.reshape(48, 50)






print (u_data_1)
print ()
print (u_data_4)



lats = np.linspace(-90, 90, 48)
pressures = np.logspace(-4, 2, 50)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(18,16), sharex=True, sharey=True)
plt.subplots_adjust(wspace=0.1, hspace=0.1)

wind_max = 5800
wind_min = -800


mp = axes[0,0].contourf(lats, pressures, u_data_1.T,  cmap=my_colors, levels=np.arange(wind_min, wind_max, 100))
cb = axes[0,0].contour(lats, pressures, u_data_1.T,
                  levels=[0],
                  colors='black', linewidths=3, linestyles='dashed')

axes[0,0].set_yscale("log")
axes[0,0].set_ylim(1e2, 1e-4)




mp = axes[0,1].contourf(lats, pressures, u_data_2.T,  cmap=my_colors, levels=np.arange(wind_min, wind_max, 100))
cb = axes[0,1].contour(lats, pressures, u_data_2.T,
                  levels=[0],
                  colors='black', linewidths=3, linestyles='dashed')

axes[0,1].set_yscale("log")
axes[0,1].set_ylim(1e2, 1e-4)




mp = axes[1,0].contourf(lats, pressures, u_data_3.T,  cmap=my_colors, levels=np.arange(wind_min, wind_max, 100))
cb = axes[1,0].contour(lats, pressures, u_data_3.T,
                  levels=[0],
                  colors='black', linewidths=3, linestyles='dashed')

axes[1,0].set_yscale("log")
axes[1,0].set_ylim(1e2, 1e-4)




mp = axes[1,1].contourf(lats, pressures, u_data_4.T,  cmap=my_colors, levels=np.arange(wind_min, wind_max, 100))
cb = axes[1,1].contour(lats, pressures, u_data_4.T,
                  levels=[0],
                  colors='black', linewidths=3, linestyles='dashed')

axes[1,1].set_yscale("log")
axes[1,1].set_ylim(1e2, 1e-4)



temp_cbar = fig.colorbar(mp, ax=axes.ravel().tolist(), location='top', aspect=50, pad=0.02)
temp_cbar.set_label('E-W Wind Speed', size=25, weight='bold')

fig.text(0.5, 0.075, r"Latitude (degrees)", size=25, weight='bold', ha='center')
fig.text(0.065, 0.5, r"Pressure (bars)", size=25, weight='bold', va='center', rotation='vertical')

plt.savefig('/home/imalsky/Desktop/winds.png', bbox_inches='tight', dpi=500)
```
